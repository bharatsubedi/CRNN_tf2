```python
import os
img_list = [file for file in os.listdir('images') if file.endswith('.png')]
print(len(img_list))
```

```python
with open('img_list.txt','w') as f:
    for i in img_list:
        a = i[:-4].split('_')
        print(a[1])
        f.write('images/{}{}{}'.format(i,' ',a[1]))
        f.write('\n')
```
